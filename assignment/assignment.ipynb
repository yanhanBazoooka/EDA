**Q1.** In class, we talked about how to compute the sample mean of a variable $X$,
$$
m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i
$$
and sample covariance of two variables $X$ and $Y$,
$$
\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))).
$$
Recall, the sample variance of $X$ is
$$
s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2.
$$
It can be very helpful to understand some basic properties of these statistics. If you want to write your calculations on a piece of paper, take a photo, and upload that to your GitHub repo, that's probably easiest.

1. Show that $m(a + bX) = a+b \times m(X)$.

\begin{aligned}
m(a + bX)
&= \frac{1}{N} \sum_{i=1}^N \bigl(a + b x_i \bigr) \\
&= \frac{1}{N} \left( \sum_{i=1}^N a + \sum_{i=1}^N b x_i \right ) \\
&= \frac{1}{N}\left(N\,a + b\,\sum_{i=1}^N x_i\right) \\
&= a + b\,\Bigl(\frac{1}{N}\sum_{i=1}^N x_i\Bigr) \\
&= a + b \, m(X).
\end{aligned}

2. Show that $\text{cov}(X,a+bY) = b \times \text{cov}(X,Y)$

\begin{aligned}
\mathrm{cov}(X,\,a + bY)
&= \frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)\,\Bigl((a + b\,y_i) - m(a + bY)\Bigr) \\
&= \frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)\,\Bigl(a + b\,y_i - \bigl(a + b\,m(Y)\bigr)\Bigr) \\
&= \frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)\,\Bigl(b\,(y_i - m(Y))\Bigr) \\
&= b\,\left[\frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)\,\bigl(y_i - m(Y)\bigr)\right] \\
&= b \,\mathrm{cov}(X,Y).
\end{aligned}


3. Show that $\text{cov}(a+bX,a+bX) = b^2 \text{cov}(X,X) $, and in particular that $\text{cov}(X,X) = s^2 $.

\begin{aligned}

\mathrm{cov}(a+bX,\,a+bX)
&= \frac{1}{N}\sum_{i=1}^N \Bigl((a + b\,x_i) - m(a + bX)\Bigr)^2 \\
&= \frac{1}{N}\sum_{i=1}^N \Bigl(b\,(x_i - m(X))\Bigr)^2 
\quad \text{(since $m(a + bX) = a + b\,m(X)$)} \\
&= b^2 \,\frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)^2 \\
&= b^2 \,\mathrm{cov}(X,X). \\

\mathrm{cov}(X,X)
&= \frac{1}{N}\sum_{i=1}^N \bigl(x_i - m(X)\bigr)^2 = s^2.\\

\end{aligned}

4. Instead of the mean, consider the median. Consider transformations that are non-decreasing (if $x\ge x'$, then $g(x)\ge g(x')$), like $2+5 \times X$ or $\text{arcsinh}(X)$. Is a non-decreasing transformation of the median the median of the transformed variable? Explain. Does your answer apply to any quantile? The IQR? The range? 

-   (a) Non-decreasing transformation of the median
Let $M = \text{median}(X)$. Because $g$ is non-decreasing, the ordering of the $x_i$ is preserved when applying $g$. Thus exactly half of the $g(x_i)$ will be below (or equal to) $g(M)$, and half above (or equal to) $g(M)$. Hence,
$$
\text{median}\bigl(g(X)\bigr) = g\bigl(\text{median}(X)\bigr).
$$

-   (b) Any quantile
A similar argument shows that for any $p$,
$$
Q_p\bigl(g(X)\bigr) = g\bigl(Q_p(X)\bigr).
$$

-   (c) IQR and range
    $$ 
    IQR(X) = Q_{0.75}(X) - Q_{0.25}(X).
    $$

    After transformation:

    $$    
    \text{IQR}\bigl(g(X)\bigr) 
    = Q_{0.75}\bigl(g(X)\bigr) - Q_{0.25}\bigl(g(X)\bigr)
    = g\bigl(Q_{0.75}(X)\bigr) - g\bigl(Q_{0.25}(X)\bigr).
    $$
    This is not simply $g(\text{IQR}(X))$ unless $g$ is a linear function.

    $$
    \text{Range}(X) = \max(X) - \min(X). Under g:
    $$

    $$
    \text{Range}\bigl(g(X)\bigr) 
    = \max\bigl(g(X)\bigr) - \min\bigl(g(X)\bigr)
    = g(\max(X)) - g(\min(X)),
    $$

    again not simply $g\bigl(\max(X) - \min(X)\bigr)$ unless $g$ is linear.


5. Consider a non-decreasing transformation $g()$. Is is always true that $m(g(X))= g(m(X))$?

$$ 
m\bigl(g(X)\bigr) 
= \frac{1}{N} \sum_{i=1}^N g(x_i),
\quad
g\bigl(m(X)\bigr) 
= g\Bigl(\tfrac{1}{N}\sum_{i=1}^N x_i\Bigr).
$$

These two expressions are not equal unless $g$ is affine. For example, for $g(x) = \log(x)$,

$$
m(\log(X)) = \frac{1}{N}\sum_{i=1}^N \log(x_i)
\quad\neq\quad
\log\Bigl(\tfrac{1}{N}\sum_{i=1}^N x_i\Bigr) = \log\bigl(m(X)\bigr).
$$

Thus for a non-decreasing but nonlinear $g$, $m\bigl(g(X)\bigr) \neq g\bigl(m(X)\bigr)$.


**Q2.** This question looks at financial transfers from foreign entities to American universities. In particular, from which countries and giftors are the gifts coming from, and to which institutions are they going? For this question, `.groupby([vars]).count()` and `.groupby([vars]).sum()` will be especially useful to tally the number of occurrences and sum the values of those occurrences.

1. Load the `./data/ForeignGifts_edu.csv` dataset.
2. For `Foreign Gift Amount`, create a histogram and describe the variable. Describe your findings.
3. For `Gift Type`, create a histogram or value counts table. What proportion of the gifts are contracts, real estate, and monetary gifts?
4. Create a kernel density plot of the log of `Foreign Gift Amount`, and then a kernel density plot of the log of `Foreign Gift Amount` conditional on gift type. Do you notice any patterns?
5. What are the top 15 countries in terms of the number of gifts? What are the top 15 countries in terms of the amount given?
6. What are the top 15 institutions in terms of the total amount of money they receive? Make a histogram of the total amount received by all institutions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
file_path = './data/ForeignGifts_edu.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset
display(df.info())
display(df.head())
```

```python
# 2. Histogram for 'Foreign Gift Amount'
plt.figure(figsize=(10,5))
sns.histplot(df['Foreign Gift Amount'].dropna(), bins=50, kde=True)
plt.xlabel('Foreign Gift Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Foreign Gift Amount')
plt.show()

description = df['Foreign Gift Amount'].describe()
display(description)
```

-   The histogram of Foreign Gift Amount is highly right-skewed, indicating that most gift amounts are relatively small, but a few extremely large values exist.

```python
# 3. Proportion of Gift Types
plt.figure(figsize=(10,5))
sns.countplot(y=df['Gift Type'], order=df['Gift Type'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Gift Type')
plt.title('Distribution of Gift Types')
plt.show()

gift_type_counts = df['Gift Type'].value_counts(normalize=True) * 100
display(gift_type_counts)
```

```python
# 4. Kernel Density Plot of Log of 'Foreign Gift Amount'
df['Log Foreign Gift Amount'] = np.log(df['Foreign Gift Amount'].replace(0, np.nan))
plt.figure(figsize=(10,5))
sns.kdeplot(df['Log Foreign Gift Amount'].dropna(), shade=True)
plt.xlabel('Log Foreign Gift Amount')
plt.ylabel('Density')
plt.title('Kernel Density Plot of Log Foreign Gift Amount')
plt.show()
```

-   The original histogram of Foreign Gift Amount was highly right-skewed.
-   Applying the log transformation makes the distribution closer to normal, which is useful for statistical analysis.

```python
# 5. Top 15 Countries by Number of Gifts
country_gift_counts = df['Country of Giftor'].value_counts().head(15)
display(country_gift_counts)
# Top 15 Countries by Amount Given
country_gift_amounts = df.groupby('Country of Giftor')['Foreign Gift Amount'].sum().nlargest(15)
display(country_gift_amounts)
```

```python
# 6. Top 15 Institutions by Total Amount Received
institution_totals = df.groupby('Institution Name')['Foreign Gift Amount'].sum().nlargest(15)
display(institution_totals)

# Histogram of Total Amount Received by All Institutions
plt.figure(figsize=(10,5))
sns.histplot(df.groupby('Institution Name')['Foreign Gift Amount'].sum(), bins=50, kde=True)
plt.xlabel('Total Foreign Gift Amount Received')
plt.ylabel('Frequency')
plt.title('Histogram of Total Foreign Gift Amount Received by Institutions')
plt.show()
```

```python
# 7. Giftors Providing the Most Money
giftor_totals = df.groupby('Giftor Name')['Foreign Gift Amount'].sum().nlargest(1)
display(giftor_totals)
```

**Q4.** This question looks at a time series of the number of active oil drilling rigs in the United States over time. The data comes from the Energy Information Agency.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.
2. To convert the `Month` variable to an ordered datetime variable, use `df['time'] = pd.to_datetime(df['Month'], format='mixed')`.
3. Let's look at `Active Well Service Rig Count (Number of Rigs)`, which is the total number of rigs over time. Make a line plot of this time series. Describe what you see.
4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.
5. The first two columns are the number of onshore and offshore rigs, respectively. Melt these columns and plot the resulting series.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("./data/drilling_rigs.csv")

# Display basic info
print("Dataset Info:")
df.info()
print("\nFirst few rows:")
print(df.head())

# Checking number of observations and variables
num_obs, num_vars = df.shape
print(f"\nNumber of observations: {num_obs}, Number of variables: {num_vars}")

# Checking for numeric columns needing coercion
print("\nColumn data types:")
print(df.dtypes)

# Convert Month to datetime
df['time'] = pd.to_datetime(df['Month'], format='mixed')

# Ensure numeric columns are correctly formatted
df_numeric_cols = df.select_dtypes(include=['object']).columns.tolist()
print("\nPotential numeric columns stored as object:", df_numeric_cols)

# Convert numerical columns stored as objects
df[df_numeric_cols] = df[df_numeric_cols].apply(pd.to_numeric, errors='coerce')

# Check again
df.info()

# Line plot of Active Well Service Rig Count
df.set_index('time', inplace=True)
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Active Well Service Rig Count (Number of Rigs)'], label='Active Rigs', color='b')
plt.xlabel("Year")
plt.ylabel("Number of Rigs")
plt.title("Active Well Service Rig Count Over Time")
plt.legend()
plt.grid()
plt.show()

# Compute first difference
df['rig_change'] = df['Active Well Service Rig Count (Number of Rigs)'].diff()

# Plot first difference
plt.figure(figsize=(12,6))
plt.plot(df.index, df['rig_change'], label='Change in Active Rigs', color='r')
plt.xlabel("Year")
plt.ylabel("Change in Number of Rigs")
plt.title("First Difference of Active Well Service Rig Count Over Time")
plt.axhline(0, linestyle='--', color='black')
plt.legend()
plt.grid()
plt.show()

# Melt Onshore and Offshore columns
melted_df = df[['Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)', 
                'Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)']]
melted_df = melted_df.reset_index()
melted_df = melted_df.melt(id_vars=['time'], var_name='Rig Type', value_name='Number of Rigs')

# Plot the melted data
plt.figure(figsize=(12,6))
sns.lineplot(data=melted_df, x='time', y='Number of Rigs', hue='Rig Type')
plt.xlabel("Year")
plt.ylabel("Number of Rigs")
plt.title("Onshore vs Offshore Rig Count Over Time")
plt.legend()
plt.grid()
plt.show()
```
